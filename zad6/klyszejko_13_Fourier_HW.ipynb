```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math
import os

# Load required files
if not os.path.exists("literki.bmp") :
    !wget https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/literki.bmp --no-check-certificate
if not os.path.exists("wzorA.bmp") :
    !wget https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/wzorA.bmp --no-check-certificate

lena_image = cv2.cvtColor(cv2.imread('lena.bmp'), cv2.COLOR_BGR2GRAY)
```

1,2,3,4 

```python
FILTER_SIZE = 21

hamming_window = np.hamming(FILTER_SIZE)
hanning_window2d = np.outer(hamming_window, hamming_window)
plt.imshow(hanning_window2d)
```

```python
FSpaceRows = 2 * np.fft.fftshift(np.fft.fftfreq(FILTER_SIZE))
FSpaceRowsM = np.outer(FSpaceRows, np.ones([1, FILTER_SIZE]))
FSpaceCols = 2 * np.fft.fftshift(np.fft.fftfreq(FILTER_SIZE))
FSpaceColsM = np.outer(np.ones([1, FILTER_SIZE]), FSpaceCols)

FreqR = np.sqrt(np.square(FSpaceRowsM) + np.square(FSpaceColsM))

plt.imshow(FreqR, interpolation="nearest", origin="upper")
plt.colorbar()
plt.show()
```

```python
FilterF = (FreqR >= 0.2) & (FreqR <= 0.3)

figFilter = plt.figure()
axsFilter = figFilter.add_subplot(projection='3d')
axsFilter.plot_surface(FSpaceRowsM, FSpaceColsM, FilterF, rstride=1, cstride=1, cmap=plt.get_cmap('gray'), linewidth=0)
figFilter.show()
```

```python
FilterFRot = np.rot90(np.fft.fftshift(np.rot90(FilterF, 2)), 2)
FilterFRot3 = np.dstack((FilterFRot, np.zeros(FilterFRot.shape)))
FilterFidft = cv2.idft(np.float32(FilterFRot3), flags=cv2.DFT_SCALE | cv2.DFT_COMPLEX_OUTPUT)
FilterFI = np.rot90(np.fft.fftshift(FilterFidft[:, :, 0]), 2)
```

```python
good_filter = hanning_window2d * FilterFI

plt.imshow(good_filter)
```

7.

```python
x_padding = lena_image.shape[0] - FILTER_SIZE
y_padding = lena_image.shape[1] - FILTER_SIZE

filter_filled = cv2.copyMakeBorder(good_filter, 0, x_padding, 0, y_padding, cv2.BORDER_CONSTANT, value=0)
```

```python
filter_filled_dft = cv2.dft(np.float32(filter_filled),flags = cv2.DFT_COMPLEX_OUTPUT)
filter_filled_shifted = np.fft.fftshift(filter_filled_dft, [0,1])
filter_filled_abs = np.abs(filter_filled_shifted)
```

8.

```python
lena_dft = cv2.dft(np.float32(lena_image),flags = cv2.DFT_COMPLEX_OUTPUT)
lena_shifted = np.fft.fftshift(lena_dft, [0, 1])

lena_filtered = lena_shifted * filter_filled_abs

lena_inv_shift = np.fft.ifftshift(lena_filtered, [0, 1])

lena_idft = cv2.idft(np.float32(lena_inv_shift) ,flags = cv2.DFT_SCALE | cv2.DFT_COMPLEX_OUTPUT)

lena_IFFT = cv2.magnitude(lena_idft[:, :, 0], lena_idft[:, :, 1])
```

```python
fig, axs = plt.subplots(1, 2)
fig.set_size_inches(10,10)

axs[0].imshow(good_filter)
axs[0].set_title('Filtr')

axs[1].imshow(lena_IFFT)
axs[1].set_title('Obraz po filtracji')
```
